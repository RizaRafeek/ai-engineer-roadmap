```python
import google.generativeai as genai 
from langchain_google_genai import ChatGoogleGenerativeAI      #To mention llm specification
from langchain.schema import HumanMessage, AIMessage
imoport re
```

```python
gemini_api_key ="...."
genai.configure(api_key = gemini_api_key)

#initialise the llm
llm = ChatGoogleGenerativeAI(
    model = :models/gemini-2.0-flash",
    google_api_key = gemini_api_key,
    temperature = 0,
    max_output_tokens = 512
)

def calculate(expression):
    return eval(expression)           #evaluate (solve) the expression and give the answer

def average_pill_weight(name):
    pill weights = {
        "Aspirin" : "500 mg",
        "Paracetamol" : "650 mg",
        "Ibuprofen" : "400 mg"
    }

known_actions = {
    "calculate" : calculate,
    "average_pill_weight" : average_pill_weight
}
```

```python
class Agent:
    def __init__(self):
        self.messages = []        #message history - maintain

    def __call__(self, message):
        self.messages.append(HumanMessage(content=message))
        response = self.execute()
        self.messages.append(AIMessage(content = response))

    def execute(self):
        response = llm(self.messages)
        return response.content

```

```python
action_re = re.compile(r'Action: \s*(\w+):\s*(.*?)(?:\n|$)', re.MULTILINE)
```

```python
def run_query(prompt, max_turns =10):
    agent = Agent()
    next_prompt = prompt

    for turn in range(max_turns):
        print(f"Turn {turn+1}:")
        response = agent(next_prompt)
        print(f"Assistant: \n{response}\n")

        #check action lines in response
        actions = action_re.findall(response)         #formatting a certain way

        if actions:
            action, action_input = actions[0]        #take the input of the first action
            action_input = action_input.strip()      #clean the input

            if action not in known_actions:
                raise ValueError(f"Unknown Action: {Action}")

            observation = known_actions[action](action_input)
            print(f"Observation: {observation}\n")
            next_prompt = f"Observation : {observation}"        #observation of frts prompt becomes the second prompt .this continues until the agent is satisfied
        else:
            if "Answer:" in response:
                print(f"Final Answer provided!")
                break
            else:
                print(f"No action found and no final answer. Stoping")
                break
    instruction_and_question = """
    You are a helpful assistant following a ReAct reasoning pattern.

    For each step:
    1.Use "Thought:" to think step by step about the  problem
    2.If you need more information, use 'Action: <action_name>:<input>' on a new line
    3.After each action. I will provide an 'Observation' with the result
    4.Continue thinking and acting until you can provide a final 'Answer:'

    Available actions:
    -calculate: perform a python calculation(e.g, calculate: 5 + 3)
    -average_pill_weight : Get the average_pill_weight of a specific pill ( e.g., average_pill_weight)

    Important : Only perform ONE action at a time, then wait for the observation

    Question : I have two pills,Aspirin and Paracetamol,
    what is their combined weight?
    """
```

```python
run_query(instruction_and_question)
```
